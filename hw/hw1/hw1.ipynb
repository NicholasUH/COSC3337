# Part 2. Data Cleaning and Preparation

```python
cars_df['car_company'] = cars_df['CarName'].apply(lambda x: x.split()[0].lower())

unique_companies = cars_df['car_company'].unique()
print(unique_companies)
```

```python
cars_df['car_company'] = cars_df['car_company'].replace('toyouta', 'toyota')
cars_df['car_company'] = cars_df['car_company'].replace('porcshce', 'porsche')
cars_df['car_company'] = cars_df['car_company'].replace('maxda', 'mazda')
cars_df['car_company'] = cars_df['car_company'].replace('vokswagen', 'volkswagen')
cars_df['car_company'] = cars_df['car_company'].replace('vw', 'Volkswagen')

print(unique_companies)
```

# Part 5. Data Pre-Processing

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

categorical_variables = ['symboling', 'fueltype', 'enginetype', 'carbody', 'doornumber', 'enginelocation'
                        ,'fuelsystem', 'cylindernumber', 'aspiration', 'drivewheel']

dummy_cars_df = pd.get_dummies(cars_df, columns=categorical_variables)


to_be_scaled = [x for x in dummy_cars_df.columns if x not in ['car_ID', 'CarName', 'car_company']]

scaled_data = StandardScaler().fit_transform(dummy_cars_df[to_be_scaled])

scaled_cars_df = pd.DataFrame(scaled_data, columns = to_be_scaled)

y = scaled_cars_df['price']
X = scaled_cars_df.drop(columns = ['price'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2024)
```

# Part 6. Model Creation and Evaluation

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score


model = LinearRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_test)

plt.scatter(y_test, predictions)
plt.xlabel('True y')
plt.ylabel('Prediction')
plt.title('True y v. Prediction')
plt.show()

residuals = y_test - predictions

sns.histplot(residuals, bins=20, kde = True)
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals')
plt.show()

r2 = r2_score(y_test, predictions)
print("R^2 score:", r2)
```

**85% of the variance can be explained by the model**

```python
coefficients = pd.DataFrame(model.coef_)
variables = pd.DataFrame(X.columns, columns=['Variable'])

result = pd.concat([variables, coefficients], axis=1)

display(result)
```

**From the coefficients, we can see the relationship between each feature and the price. If positive, we can see their is a positive linear relationship, if negative, its a negative linear relationship. We can also see the strength of each relationship based on the value.**
