```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris_df = pd.read_csv('iris-1.csv')
iris_df.head()
```

```python
iris_df.plot(kind = 'scatter', x = 'petal_width', y = 'sepal_length', figsize=(6,4))
```

```python
petal_width = iris_df['petal_width']
a = np.ones((petal_width.shape[0], 1))
X = np.hstack((a,petal_width.values.reshape(-1,1)))
y = iris_df['sepal_length']
```

```python
theta = np.matmul(np.linalg.inv(np.matmul(X.T,X)), np.matmul(X.T,y))
theta
```

```python
iris_df.plot(kind = 'scatter', x = 'petal_width', y = 'sepal_length', figsize=(6,4))
plt.axis([0,3,4,10])
X = np.linspace(0,3)
plt.plot(X,theta[0] + theta[1] * X, "b") #plot the model here
plt.text(0.5, 8, r"$\theta_0 = 4.78$", fontsize=13, color="b") #display theta_0 here as shown below
plt.text(0.5, 7, r"$\theta_1 = .89$", fontsize=13, color="b") #display theta_1 here as shown below
plt.show()
```

```python
example_petal_width = 1.37
predicted_sepal_length = theta[0] * theta[1] * example_petal_width
print('The predicted sepal_length is:', predicted_sepal_length)
```

```python
iris_df.plot(kind = 'scatter', x = 'petal_width', y = 'sepal_length', figsize=(6,4))
plt.axis([0,3,0,10])
X = np.linspace(0,3)
plt.plot(X,theta[0] + theta[1] * X, "b") #plot the model here
plt.text(0.5, 8, r"$\theta_0 = 4.78$", fontsize=13, color="b") #display theta_0 here as shown below
plt.text(0.5, 7, r"$\theta_1 = .89$", fontsize=13, color="b") #display theta_1 here as shown below
plt.plot([example_petal_width,example_petal_width],[0,predicted_sepal_length],"r--")
plt.text(1.5, 4, r"Prediction = 5.81", fontsize=13, color="b") #display your prediction here as shown below
plt.plot(example_petal_width,predicted_sepal_length,"ro") #plot the petal width with respect to its predicted sepal length
plt.show()
```
