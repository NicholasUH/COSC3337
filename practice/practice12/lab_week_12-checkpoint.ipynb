```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from matplotlib import rcParams
rcParams['figure.figsize'] = 15, 5
sns.set_style('darkgrid')
```

```python
new_customer_df = pd.read_csv('customers.csv')
new_customer_df.drop(['CustomerID'], axis=1, inplace=True)
new_customer_df.head()
```

```python
new_customer_df.info()
```

```python
new_customer_df_hmap = new_customer_df.copy()
non_numeric_columns = new_customer_df_hmap.select_dtypes(exclude=['number']).columns
new_customer_df_hmap[non_numeric_columns] = new_customer_df_hmap[non_numeric_columns].apply(pd.to_numeric, errors='coerce')

sns.heatmap(new_customer_df_hmap.corr(), annot=True)
plt.show()
```

```python
male_count = (new_customer_df['Genre'] == 'Male').sum()
female_count = (new_customer_df['Genre'] == 'Female').sum()
gender_counts = {'Male': male_count, 'Female': female_count}

labels = gender_counts.keys()
sizes = gender_counts.values()
explode = (0, 0.1)
colors = ['skyblue', 'pink']

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', colors=colors)
ax1.axis('equal')

plt.title('Proportions of Male to Female')
plt.show()
```

```python
X = new_customer_df.iloc[:, [2, 3]].values
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=3, min_samples=4, metric='euclidean')
db.fit(X)
```

```python
db.labels_
```

```python
y_preds = db.fit_predict(X)

plt.scatter(X[y_preds == 0, 0], X[y_preds == 0, 1], s = 50, c = 'pink', label='cluster 1')
plt.scatter(X[y_preds == 1, 0], X[y_preds == 1, 1], s = 50, c = 'yellow', label='cluster 2')
plt.scatter(X[y_preds == 2, 0], X[y_preds == 2, 1], s = 50, c = 'cyan', label='cluster 3')
plt.scatter(X[y_preds == 3, 0], X[y_preds == 3, 1], s = 50, c = 'magenta', label='cluster 4')
plt.scatter(X[y_preds == 4, 0], X[y_preds == 4, 1], s = 50, c = 'orange', label='cluster 5')
plt.scatter(X[y_preds == 5, 0], X[y_preds == 5, 1], s = 50, c = 'blue', label='cluster 6')
plt.scatter(X[y_preds == 6, 0], X[y_preds == 6, 1], s = 50, c = 'red', label='cluster 7')
plt.scatter(X[y_preds == 7, 0], X[y_preds == 7, 1], s = 50, c = 'black', label='cluster 8')
plt.scatter(X[y_preds == 8, 0], X[y_preds == 8, 1], s = 50, c = 'violet', label='cluster 9')
plt.title('Clusters of customers')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
