# Exploratory Data Analysis

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
from matplotlib import rcParams 
rcParams['figure.figsize'] = 15, 5 
sns.set_style('darkgrid')
```

```python
new_diabetes_df = pd.read_csv('diabetes.csv') 
new_diabetes_df.head()
```

```python
new_diabetes_df.info()
```

```python
new_diabetes_df.describe()
```

```python
new_diabetes_df_copy = new_diabetes_df.copy(deep=True) 
new_diabetes_df_copy['Glucose'] = new_diabetes_df_copy['Glucose'].replace(0,np.NaN) 
new_diabetes_df_copy['BloodPressure'] = new_diabetes_df_copy['BloodPressure'].replace(0,np.NaN) 
new_diabetes_df_copy['SkinThickness'] = new_diabetes_df_copy['SkinThickness'].replace(0,np.NaN) 
new_diabetes_df_copy['Insulin'] = new_diabetes_df_copy['Insulin'].replace(0,np.NaN) 
new_diabetes_df_copy['BMI'] = new_diabetes_df_copy['BMI'].replace(0,np.NaN)
```

```python
new_diabetes_df_copy[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']].hist(figsize = (20,10)) 
plt.show()
```

```python
new_diabetes_df_copy['Glucose'].fillna(new_diabetes_df_copy['Glucose'].mean(), inplace = True) 
new_diabetes_df_copy['BloodPressure'].fillna(new_diabetes_df_copy['BloodPressure'].mean(), inplace = True) 
new_diabetes_df_copy['SkinThickness'].fillna(new_diabetes_df_copy['SkinThickness'].median(), inplace = True) 
new_diabetes_df_copy['Insulin'].fillna(new_diabetes_df_copy['Insulin'].median(), inplace = True) 
new_diabetes_df_copy['BMI'].fillna(new_diabetes_df_copy['BMI'].median(), inplace = True)
```

```python
sns.heatmap(new_diabetes_df_copy.corr(), annot=True) 
plt.title('Correlation Matrix') 
plt.show() 
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5)) 
 
sns.scatterplot(x='BloodPressure', y='BMI', data=new_diabetes_df_copy, alpha=0.3, ax=axes[0]) 
axes[0].set_title('BloodPressure VS. BMI') 
 
sns.scatterplot(x='Glucose', y='DiabetesPedigreeFunction', data=new_diabetes_df_copy, alpha=0.3, ax=axes[1]) 
axes[1].set_title('Glucose VS. DPF') 
 
plt.show() 
```

```python
import statsmodels.api as sm 
import scipy 
import pylab 
 
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5)) 
 
sns.histplot(new_diabetes_df_copy['Glucose'], ax=axes[0]) 
axes[0].set_title('Glucose Distribution') 
 
sm.qqplot(new_diabetes_df_copy['Glucose'], line='s', ax=axes[1]) 
axes[1].set_title('Glucose Q-Q Plot') 
 
pylab.show() 
 
w, p_val = scipy.stats.shapiro(new_diabetes_df_copy['Glucose']) 
print('Shapiro-Wilk: \nw:{}, p-value:{}\n'.format(w,p_val)) 
 
d, p_val = scipy.stats.kstest(new_diabetes_df_copy['Glucose'], 'norm') 
print('Kolmogorov-Smirnov: \nd:{}, p-value:{}\n'.format(w,p_val)) 
 
print('Skewness of the data: \n{}\n'.format(scipy.stats.skew(new_diabetes_df_copy['Glucose'])))
```

```python
new_class_zero =new_diabetes_df_copy[(new_diabetes_df_copy['Outcome'] == 0)] 
new_class_one = new_diabetes_df_copy[(new_diabetes_df_copy['Outcome'] == 1)] 
 
plt.hist(x=new_class_zero['Glucose'], label='class 0', alpha=0.5) 
plt.hist(x=new_class_one['Glucose'], label='class 1', alpha=0.5) 
plt.legend() 
plt.title('Glucose Distribution') 
 
plt.show()
```

# Pre-Processing

```python
from sklearn.preprocessing import StandardScaler 
 
X = new_diabetes_df_copy.drop('Outcome', axis=1) 
y = new_diabetes_df_copy['Outcome'] 
 
scaler = StandardScaler() 
X_scaled = scaler.fit_transform(X) 
X = pd.DataFrame(data=X_scaled, columns=X.columns) 
```

```python
X.head()
```

# Creating our Model

```python
sns.countplot(x=new_diabetes_df_copy['Outcome']) 
plt.show() 
```

```python
from sklearn.model_selection import train_test_split 
 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=100)
```

```python
from sklearn.neighbors import KNeighborsClassifier 
 
test_scores = [] 
train_scores = [] 
 
for i in range(1,15): 
    knn = KNeighborsClassifier(i) 

    knn.fit(X_train,y_train) 
     
    train_scores.append(knn.score(X_train,y_train)) 
    test_scores.append(knn.score(X_test,y_test)) 
```

```python
sns.lineplot(x=range(1,15), y=train_scores, marker='*', label='Train Score') 
sns.lineplot(x=range(1,15), y=test_scores, marker='o', label='Test Score') 
plt.title('K vs. Score') 
plt.xlabel('K') 
plt.ylabel('Score') 
plt.show()
```

```python
knn = KNeighborsClassifier(11) 
 
knn.fit(X_train,y_train) 
knn.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import classification_report 
 
y_pred = knn.predict(X_test) 
print(classification_report(y_test,y_pred)) 
print(confusion_matrix(y_test,y_pred))
```
